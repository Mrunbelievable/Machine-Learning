```python
#problem 3
# part 1----------------------------------------------------
# the probability of class y
# p(y=-1) = 6/10,  P(y=1) = 4/10
# Finding all the individual feature probabilities p(xi|yi)
# First choose y=-1, we have the probablity below:
# P(x1=1|y=-1) = 3/6, P(x1=0|y=-1) = 3/6
# P(x2=1|y=-1) = 5/6, P(x2=0|y=-1) = 1/6
# P(x3=1|y=-1) = 4/6, P(x3=0|y=-1) = 2/6
# P(x4=1|y=-1) = 5/6, P(x4=0|y=-1) = 1/6
# P(x5=1|y=-1) = 2/6, P(x5=0|y=-1) = 4/6
# Now, let y=1, we have
# P(x1=1|y=1) = 3/4, P(x1=0|y=1) = 1/4
# P(x2=1|y=1) = 0/4, P(x2=0|y=1) = 4/4
# P(x3=1|y=1) = 3/4, P(x3=0|y=1) = 1/4
# P(x4=1|y=1) = 2/4, P(x4=0|y=1) = 2/4
# P(x5=1|y=1) = 1/4, P(x5=0|y=1) = 3/4
# part 2----------------------------------------------------
# predicted for X(00000)
# When class y=1
# P(y=1,X1=X2=X3=X4=X5=0) = P(y=1)*P(x1=0|y=1)*P(x2=0|y=1)*P(x3=0|y=1)*P(x4=0|y=1)*P(x5=0|y=1)
# P(y=1,X1=X2=X3=X4=X5=0) = 0.0094
# When class y=-1
# P(y=-1,X1=X2=X3=X4=X5=0) = P(y=1)*P(x1=0|y=-1)*P(x2=0|y=-1)*P(x3=0|y=-1)*P(x4=0|y=-1)*P(x5=0|y=-1)
# P(y=-1,X1=X2=X3=X4=X5=0) = 0.0019
# according to the result we have, class y = 1 would be predicted for x=(00000)
# predicted for X(11010)
# When class y=1
# P(y=1,X=11010) = P(y=1)*P(x1=1|y=1)*P(x2=1|y=1)*P(x3=0|y=1)*P(x4=1|y=1)*P(x5=0|y=1)
# P(y=1,X=11010) = 0
# When class y=-1
# P(y=-1,X=11010) = P(y=-1)*P(x1=1|y=-1)*P(x2=1|y=-1)*P(x3=0|y=-1)*P(x4=1|y=-1)*P(x5=0|y=-1)
# P(y=-1,X=11010) = 0.0463
# according to the result we have, class y = -1 would be predicted for x=(11010)
# Part 3----------------------------------------------------
# P(y=1|x=00000) = p(y=1,X=00000)/(p(y=1,X=00000)+p(y=-1,X=00000)) = 0.83
# P(y=1|x=11010) = p(y=1,X=11010)/(p(y=1,X=11010)+p(y=-1,X=11010)) = 0
# Part 4----------------------------------------------------
# Reason why we use naive Bayes Classifier is that we assume each feature are independent
# If we use join Bayes Classifier, it will become more complexity
# It will have 2^5-1 probabilities to be calculated
# In this process of join Bayes, the result will be overfitting
# Part 5----------------------------------------------------
# We do not need to re-train the model
# P(y=1|x2,x3,x4,x5) = (P(x2,x3,x4,x5|y=1)*P(y=1))/(P(x2,x3,x4,x5|y=1)*P(y=1)+P(x2,x3,x4,x5|y=-1)*P(y=-1))
```
